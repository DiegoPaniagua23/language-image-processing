````python
# Dataset: puntos representando (edad, porcentaje de personas con diabetes)
# Cada tupla es (edad promedio, % de personas con diabetes en ese grupo)
points = [(20, 0.6),
          (30, 3.2),
          (40, 8.7),
          (50, 18.0),
          (60, 25.8),
          (70, 24.9)]
````

````python
# Funci√≥n de p√©rdida cuadr√°tica: mide el error total al ajustar una l√≠nea y = wx
def F(w):
    # points es una lista de tuplas (x, y)
    # Se calcula la suma de los errores cuadrados: (wx - y)^2
    return sum((w * x - y)**2 for x, y in points)

# Derivada de la funci√≥n de p√©rdida respecto a w (gradiente)
def dF(w):
    # Derivada de (wx - y)^2 con respecto a w es: 2 * (wx - y) * x
    # Esto se usa para encontrar el m√≠nimo de F(w) (por ejemplo, usando gradiente descendente)
    return sum(2 * (w * x - y) * x for x, y in points)
````

### üîß ¬øQu√© es `eta`?

```python
eta = 0.01
```

- `eta` es el **learning rate** o **tasa de aprendizaje**.
- Controla **qu√© tan grande es el paso** que damos en cada iteraci√≥n para actualizar el par√°metro `w`.
- Es un n√∫mero peque√±o positivo.

---

### üìâ ¬øC√≥mo afecta a la actualizaci√≥n de `w`?

En el gradiente descendente, actualizamos el par√°metro as√≠:

$
w_{\text{nuevo}} = w_{\text{viejo}} - \eta \cdot \nabla F(w)$


- Si `eta` es **muy grande**, el algoritmo puede **saltar** el m√≠nimo y **no converger**.
- Si `eta` es **muy peque√±o**, el algoritmo avanza muy lento y puede tardar en converger.

---

### ‚ö†Ô∏è Ejemplo intuitivo

Imagina que est√°s bajando una colina (funci√≥n de p√©rdida):

- `gradient` te dice **la pendiente** (hacia d√≥nde y cu√°nto bajar).
- `eta` te dice **qu√© tan grande es tu paso** hacia abajo.

Un `eta` apropiado te ayuda a **llegar al m√≠nimo de la colina r√°pido y sin pasarte**.

---

### ‚úÖ En resumen:

- `eta` regula la **velocidad de aprendizaje**.
- Es un par√°metro que puede influir mucho en la **eficacia y estabilidad** del entrenamiento.
- En tu c√≥digo, `eta = 0.01` indica que est√°s dando **pasos peque√±os y seguros** para encontrar el mejor valor de `w`.

Veamos la herramienta: https://uclaacm.github.io/gradient-descent-visualiser/#playground


````python
# Inicializamos el par√°metro w (pendiente de la recta) en 0
w = 0

# Tasa de aprendizaje (learning rate), controla qu√© tan grandes son los pasos del gradiente
eta = 0.001

# Iteramos 150 veces para ajustar w mediante gradiente descendente
for t in range(100):
    # Evaluamos el valor actual de la funci√≥n de p√©rdida F(w)
    value = F(w)

    # Calculamos el gradiente (derivada de F respecto a w)
    gradient = dF(w)

    # Guardamos el valor anterior de w (para impresi√≥n)
    wo = w

    # Actualizamos w en direcci√≥n opuesta al gradiente (para minimizar la funci√≥n)
    w = w - eta * gradient

    # Mostramos el progreso de cada iteraci√≥n
    print('Iteration {}: wo = {}, w = {}, F(w) = {}, Gradiente = {}'.format(
        t, wo, w, value, gradient
    ))



````

````python
xAxis=range(0,100);
pyplot.plot(xAxis, [w*i for i in xAxis]);
for x, y in points:
    pyplot.plot(x,y, marker='o', color="red");
pyplot.show()
````
