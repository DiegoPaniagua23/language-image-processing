````python
# Dataset: puntos representando (edad, porcentaje de personas con diabetes)
# Cada tupla es (edad promedio, % de personas con diabetes en ese grupo)
points = [(20, 0.6),
          (30, 3.2),
          (40, 8.7),
          (50, 18.0),
          (60, 25.8),
          (70, 24.9)]
````

````python
# Función de pérdida cuadrática: mide el error total al ajustar una línea y = wx
def F(w):
    # points es una lista de tuplas (x, y)
    # Se calcula la suma de los errores cuadrados: (wx - y)^2
    return sum((w * x - y)**2 for x, y in points)

# Derivada de la función de pérdida respecto a w (gradiente)
def dF(w):
    # Derivada de (wx - y)^2 con respecto a w es: 2 * (wx - y) * x
    # Esto se usa para encontrar el mínimo de F(w) (por ejemplo, usando gradiente descendente)
    return sum(2 * (w * x - y) * x for x, y in points)
````

### 🔧 ¿Qué es `eta`?

```python
eta = 0.01
```

- `eta` es el **learning rate** o **tasa de aprendizaje**.
- Controla **qué tan grande es el paso** que damos en cada iteración para actualizar el parámetro `w`.
- Es un número pequeño positivo.

---

### 📉 ¿Cómo afecta a la actualización de `w`?

En el gradiente descendente, actualizamos el parámetro así:

$
w_{\text{nuevo}} = w_{\text{viejo}} - \eta \cdot \nabla F(w)$


- Si `eta` es **muy grande**, el algoritmo puede **saltar** el mínimo y **no converger**.
- Si `eta` es **muy pequeño**, el algoritmo avanza muy lento y puede tardar en converger.

---

### ⚠️ Ejemplo intuitivo

Imagina que estás bajando una colina (función de pérdida):

- `gradient` te dice **la pendiente** (hacia dónde y cuánto bajar).
- `eta` te dice **qué tan grande es tu paso** hacia abajo.

Un `eta` apropiado te ayuda a **llegar al mínimo de la colina rápido y sin pasarte**.

---

### ✅ En resumen:

- `eta` regula la **velocidad de aprendizaje**.
- Es un parámetro que puede influir mucho en la **eficacia y estabilidad** del entrenamiento.
- En tu código, `eta = 0.01` indica que estás dando **pasos pequeños y seguros** para encontrar el mejor valor de `w`.

Veamos la herramienta: https://uclaacm.github.io/gradient-descent-visualiser/#playground


````python
# Inicializamos el parámetro w (pendiente de la recta) en 0
w = 0

# Tasa de aprendizaje (learning rate), controla qué tan grandes son los pasos del gradiente
eta = 0.001

# Iteramos 150 veces para ajustar w mediante gradiente descendente
for t in range(100):
    # Evaluamos el valor actual de la función de pérdida F(w)
    value = F(w)

    # Calculamos el gradiente (derivada de F respecto a w)
    gradient = dF(w)

    # Guardamos el valor anterior de w (para impresión)
    wo = w

    # Actualizamos w en dirección opuesta al gradiente (para minimizar la función)
    w = w - eta * gradient

    # Mostramos el progreso de cada iteración
    print('Iteration {}: wo = {}, w = {}, F(w) = {}, Gradiente = {}'.format(
        t, wo, w, value, gradient
    ))



````

````python
xAxis=range(0,100);
pyplot.plot(xAxis, [w*i for i in xAxis]);
for x, y in points:
    pyplot.plot(x,y, marker='o', color="red");
pyplot.show()
````
