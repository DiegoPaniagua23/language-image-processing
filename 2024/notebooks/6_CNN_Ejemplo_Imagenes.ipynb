```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

# Cargar la imagen
imagen = cv2.imread(imagen)
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
```

```python
# Matriz de convolución para detección de bordes (filtro Laplaciano)
kernel_bordes = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

kernel_sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

kernel_sobel_y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
])

kernel_solo_2 = np.array([
    [2, 2, 2],
    [ 2,  2,  2],
    [ 2,  2,  2]
])

kernel_solo_menos_2 = np.array([
    [ -2, -2, -2],
    [ -2,  0, -2],
    [ -2, -2, -2]
])


```

```python
# Generar matriz aleatoria de 3x3 con valores entre -2 y 2
kernel_aleatorio = np.random.randint(-2, 3, size=(3, 3))  # el límite superior en randint es exclusivo
print("Matriz aleatoria:")
print(kernel_aleatorio)


kernel = kernel_aleatorio
kernel
```

```python
# Aplicar convolución (filtro)
bordes = cv2.filter2D(gris, -1, kernel)

# Mostrar resultados
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bordes, cmap='gray')
plt.title('Bordes con convolución')
plt.axis('off')

plt.tight_layout()
plt.show()
```
